```python
def type_hierarchy(type):
    if re.match("^symbolic link to", type):
        return "Symbolic link"
    if re.match("^broken symbolic link to", type):
        return "Broken symbolic link"
    elif re.match("ELF (64|32).bit", type):
        return "Shared library"
    elif re.match("^GNU message catalog", type):
        return "GNU message catalog"
    elif re.match("^(.* image data|X pixmap image|MS Windows icon resource|SVG Scalable Vector Graphics image)", type):
        return "Image"
    elif re.match("^a .*python[1-9]? script", type) or re.match("^(P|p)ython script", type):
        return "Python script"
    elif re.match("^(ASCII text|HTML|LaTeX|TeX|OpenDocument Text|PDF document|PostScript document|exported SGML document|Microsoft Excel 2007|Composite Document File V2 Document)", type):
        return "Document"
    elif re.match("^((C|C\+\+|Pascal) source|M4 macro processor script)", type):
        return "Source code"
    elif re.match("^((bzip2|gzip|XZ) compressed data|Zip archive data|current ar archive)", type):
        return "Compressed file"
    elif re.match("^(Web Open Font Format|OpenType font data|Embedded OpenType|TrueType font data|ASCII font metrics)", type):
        return "Font"
    elif re.match("^GNU gettext message catalogue|GNU message catalog|timezone data|SQLite 3.x database", type):
        return "Data"
    elif re.match("^(PE32|COM executable for DOS)", type):
        return "Windows executable"
    elif re.match("^(.+Unicode|ISO.+) text", type):
        return "Text"
    elif re.match("^unified diff output|troff or preprocessor input|Unicode text", type):
        return "Text"
    elif re.match("^(data|FoxPro FPT|X1 archive data|SAS 7\+ data file|Matlab v5 mat\-file|XML|RIFF|Hierarchical Data Format|NetCDF Data Format data)", type):
        return "Data"
    elif re.match("^(POSIX shell script|Bourne-Again shell script|a .*env sh script)", type):
        return "Shell script"
    elif re.match("^(Perl script)", type):
        return "Perl script"
    elif re.match("^(FORTRAN program)", type):
        return "Executable"
    elif re.match("^((automake )?makefile script)", type):
        return "Make"
    elif re.match("^(^python .* byte)", type):
        return "Python byte code"
    elif re.match("^DOS batch file", type):
        return "DOS batch file"
    elif re.match("^Palm OS dynamic library data", type):
        return "Palm OS dynamic library data"
    
    else:
        return type
```

```python
def md5(fname):
    hash_md5 = hashlib.md5()
    with open(fname, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()
```
